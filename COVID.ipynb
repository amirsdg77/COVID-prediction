# ZScore

```python
from scipy import stats
```

```python
z = np.abs(stats.zscore(df1))
z
```

```python
print(np.where(z > 3))
```

```python
ppd = df1[(z<3).all(axis=1)]
ppd.shape
```

```python
ppd
```

# IQR

```python
Q1 = df1.quantile(0.25)
Q1
```

```python
Q3 = df1.quantile(0.75)
Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
IPPD = df1[((df1 >= (Q1 - 1.5*IQR)) & (df1 <= (Q3 + 1.5*IQR))).all(axis = 1)]
IPPD
```

# Multiple linear regression

# the ZScore deleted outlier

```python
ppd.head()
```

```python
df2 = new_df.iloc[:,3:6]
new_df = ppd.join(df2)
new_df
```

```python
y = y.transpose()
y = y.drop(columns = 0, axis = 1)
y = y.drop(columns = 14, axis = 1)
y = y.transpose()
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(new_df, y, test_size = 0.3, random_state = 101)
```

```python
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

```python
lr.intercept_
```

```python
prediction_train = lr.predict(X_train)
plt.scatter(Y_train, prediction_train)
```

```python
prediction_test = lr.predict(X_test)
plt.scatter(Y_test, prediction_test)
```

```python
from sklearn import metrics
```

```python
print('RMSE train', np.sqrt(metrics.mean_squared_error(Y_train, prediction_train)))
print('MSE train', metrics.mean_squared_error(Y_train, prediction_train))
print('RMSE test', np.sqrt(metrics.mean_squared_error(Y_test, prediction_test)))
print('MSE test', metrics.mean_squared_error(Y_test, prediction_test))
```
